## Load Foto & Konversi RGB

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

path_backlight = 'backlight.jpg' 
img_backlight_bgr = cv2.imread(path_backlight)

img_backlight_rgb = cv2.cvtColor(img_backlight_bgr, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(8, 6))
plt.imshow(img_backlight_rgb)
plt.title('Foto Asli (Backlight)')
plt.axis('off')
plt.show()
```

## Grayscale

```python
img_gray = cv2.cvtColor(img_backlight_rgb, cv2.COLOR_RGB2GRAY)

plt.figure(figsize=(8, 6))
plt.imshow(img_gray, cmap='gray')
plt.title('1. Hasil Konversi Grayscale')
plt.axis('off')
plt.show()
```

## Brightness

```python
nilai_brightness = 70 

img_bright = cv2.convertScaleAbs(img_gray, alpha=1.0, beta=nilai_brightness)

plt.figure(figsize=(8, 6))
plt.imshow(img_bright, cmap='gray')
plt.title(f'2. Hasil Penyesuaian Brightness (Beta={nilai_brightness})')
plt.axis('off')
plt.show()
```

## Kontras

```python
nilai_contrast = 1.5 

img_final = cv2.convertScaleAbs(img_bright, alpha=nilai_contrast, beta=0)

plt.figure(figsize=(8, 6))
plt.imshow(img_final, cmap='gray')
plt.title(f'3. Hasil Final Penyesuaian Kontras (Alpha={nilai_contrast})')
plt.axis('off')
plt.show()
```

## Percerah & Kontras

```python
img_final = cv2.convertScaleAbs(img_bright, alpha=nilai_contrast, beta=0)

plt.figure(figsize=(8, 6))
plt.imshow(img_final, cmap='gray')
plt.title('Gambar Gray yang Dipercerah dan Diperkontras')
plt.axis('off')
plt.show()
```
